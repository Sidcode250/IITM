```python
# generating random outcome
import numpy as np

def uniform(n,m):
  return np.random.randint(1, n+1, size = m)
```

```python
# tossing a coin
print(uniform(2,10))

# throwing a dice
print(uniform(6,10)) # 6 outcomes, 10 times dice thrown
```

```python
# Monte Carlo
# P(a) ~ Na / N => Na = no. of times success, N = no. of times tested

#tossing a coin
count = 0
for i in range(100): #tossing 100 times
  if uniform(2,1) == 1: #checking for success
    count += 1
print("P(a) = ", count/100)
```

```python
#Monte hall problem
#to switch doors or not

no = 0
for i in range(100):
  car_loc = uniform(3,1)

  if car_loc == 1:
    goat_1_loc = 2; goat_2_loc = 3
  elif car_loc == 2:
    goat_1_loc = 1; goat_2_loc = 3
  else:
    goat_1_loc = 1; goat_2_loc = 2

  user_choice = uniform(3,1)
  if user_choice == goat_1_loc:
    host_show = goat_2_loc; other_choice = car_loc
  elif user_choice == goat_2_loc:
    host_show = goat_1_loc; other_choice = car_loc
  else:
    host_show = goat_1_loc; other_choice = goat_2_loc

  if other_choice == car_loc:
    no += 1

print(no/100)
#switch
```

```python
# Gambling Problem

# 0 < k < N
# he has k amount of money, if he gets to N he leaves happy, if he gets to 0 he leaves bankrupt
# We have to check the probability of him going bankrupt

no1 = 0
k = 5; N = 10
for i in range(100):
  k = 5
  while k > 0 and k < N:
    if uniform(2,1 )== 1:
      k = k + 1
    else:
      k = k - 1

  if k == 0:
    no1 = no1 + 1

print(no1/100)
```
